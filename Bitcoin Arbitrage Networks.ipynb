```python
import requests
import networkx as nx
```

```python
fx_id = '' #get api key here - https://openexchangerates.org/signup
```

```python
url = 'http://openexchangerates.org/api/historical/%s.json?app_id=%s' % ('2014-09-06', fx_id)
```

```python
r = requests.get(url)
```

```python
usd_rates = r.json()['rates']
usd_rates
```

```python
btc_price_url = 'http://api.bitcoincharts.com/v1/weighted_prices.json'
```

```python
r2 = requests.get(btc_price_url)
```

```python
btc_dict = r2.json()
btc_rates = {}
for k in btc_dict:
    try:
        btc_rates[k] = float(btc_dict[k]['24h'])
    except:
        pass
```

```python
btc_rates
```

```python
D = nx.DiGraph()
```

```python
def add_fx_rates(D, base_curr, fx_rates):
    D.add_node(base_curr)
    for curr, fxr in fx_rates.items():
        D.add_node(curr.encode('ascii','ignore'))
        #fxr = float(fxr)
        D.add_edge(base_curr, curr, weight = fxr)
        D.add_edge(curr, base_curr, weight = 1/fxr)
    return D
```

```python
D = add_fx_rates(D, 'USD', usd_rates)
```

```python
D = add_fx_rates(D, 'BTC', btc_rates)
```

```python
def edge_chain_product(D, c):
    n = 0
    val = 1
    while n < len(c)-1:
        val *= float(D.edge[c[n]][c[n+1]]['weight'])
        n+=1
    val *= float(D.edge[c[n]][c[0]]['weight'])
    return val
```

```python
def find_cycle_returns(D, verbose = False):
    fx_cycles = {}
    max_return = 1.0
    max_cycle = 'NA'
    for c in nx.simple_cycles(D):
        if len(c) > 2:
            val = edge_chain_product(D, c)
            fx_cycles[tuple(c)] = val
            if verbose:
                print_returns(val, c)
            if val > max_return:
                max_return = val
                max_cycle = c
    return fx_cycles, max_cycle, max_return
```

```python
cr, max_cr, max_r = find_cycle_returns(D)
```

```python
def print_returns(rr, c):
    print 'You could earn a %.2f percent rate of return trading this group of currencies: %s' % ((rr-1)*100, c)
```

```python
print_returns(max_r, max_cr)
```
